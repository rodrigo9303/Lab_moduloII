```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
circle_x = 5.
circle_y = 5.
L=6.
N=1000
phi = 2. * np.pi * np.random.rand(N)
```

```python
ranillo = L*np.random.uniform(1.,1.5,N)
rani = np.sqrt(ranillo)
x = rani * np.cos(phi) + circle_x
y = rani * np.sin(phi) + circle_y
plt.plot(x,y,'.')
plt.xlabel('x')
plt.ylabel('y')
```

```python
def distancia(x1,x2,y1,y2):#se define el como se va a obtener la distancia
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dis_datos_ani = [] # se almacenan los datos de todas las distancias
for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(x[i],x[j],y[i],y[j])
        dis_datos_ani.append(d_d)
```

```python
minimos=[] # En esta lista se almacenarán todas las distancias minimas
for i in range(N-1):
    dista = [] # En esta lsita se tienen todas las distancias para cada uno de los puntos pivote (999) y que despues se pide la distancia minima.
    for j in range(i+1,N):
        d_dc = distancia(x[i],x[j],y[i],y[j])
        dista.append(d_dc)    
    mi=min(dista) # aquí se pide la distancia minima para cada punto pivote, que después se almacenará en la lista minimos.
    minimos.append(mi) 
#    print('minimo = ',mi)
print(len(minimos)) 
```

```python
velocidad = [] # Se crea una velocidad en la direccion del punto más cercano al punto pivote. v=distancia_minima*0.1
for a in minimos:
    velocidad.append(a*0.1)
print(len(velocidad))
```

```python
y_new1 = []  #se creea el nuevo punto.
for a,b in enumerate(velocidad):   
    y_new1.append(y[a]+velocidad[a])
```
